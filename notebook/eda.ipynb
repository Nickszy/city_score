# 本文件为对单项指标数据的探索

```python
import pandas as pd
import numpy as np
import os
```

```python
base = '../长三角城市年度数据/'
files = os.listdir(base)
```

```python
files
```

```python
data = pd.read_excel(base+files[3])
localname = data.iloc[0,1]
data.drop(index=[0,1,2],inplace=True)
data.index=data['指标']
data.index.name='年份'
data.drop(columns=['指标'],inplace=True)
data.columns.name='指标'
# data = data.convert_dtypes()
data.describe()
```

```python
data.columns.tolist()
```

```python
from matplotlib.pyplot import plot as  plt
edudata = data[[ '一般公共预算支出_教育_市辖区','城镇单位就业人员数_教育_市辖区','普通中学学校数_市辖区','普通小学学校数_市辖区','普通高等学校专任教师数_市辖区','普通中学专任教师数_市辖区','普通小学专任教师数_市辖区','普通高等学校在校生数_市辖区','普通中学在校生数_市辖区','普通小学在校生数_市辖区','公共图书馆总藏量_市辖区','中等职业教育学校数_市辖区','中等职业教育专任教师数_市辖区','中等职业教育在校生数_市辖区']]
# edudata = edudata.dropna()
edudata = pd.DataFrame(edudata,dtype=np.float)
edudata
```

```python
# 标准化
for i in edudata.columns:
    edudata[i]=round((edudata[i]-np.mean(edudata[i]))/(np.max(edudata[i])-np.min(edudata[i])),3)
edudata.plot(figsize=(15,10))
```

```python
edudata.head()
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts
def line_base(data):
    c = (
        Line()
        .add_xaxis(data.index.tolist())
        .add_yaxis('一般公共预算支出_教育_市辖区',data['一般公共预算支出_教育_市辖区'].tolist())
        .add_yaxis('公共图书馆总藏量_市辖区',data['公共图书馆总藏量_市辖区'].tolist())
        .add_yaxis('普通小学学校数_市辖区',data['普通小学学校数_市辖区'].tolist())
        .add_yaxis('普通小学专任教师数_市辖区',data['普通小学专任教师数_市辖区'].tolist())
        .add_yaxis('普通小学在校生数_市辖区',data['普通小学在校生数_市辖区'].tolist())
        .add_yaxis('普通中学专任教师数_市辖区',data['普通中学专任教师数_市辖区'].tolist())
        .add_yaxis('普通中学学校数_市辖区',data['普通中学学校数_市辖区'].tolist())
        .add_yaxis('普通中学在校生数_市辖区',data['普通中学在校生数_市辖区'].tolist())
        .add_yaxis('普通高等学校在校生数_市辖区',data['普通高等学校在校生数_市辖区'].tolist())
        .add_yaxis('普通高等学校专任教师数_市辖区',data['普通高等学校专任教师数_市辖区'].tolist())
        .add_yaxis('中等职业教育学校数_市辖区',data['中等职业教育学校数_市辖区'].tolist(),symbol='triangle')
        .add_yaxis('中等职业教育专任教师数_市辖区',data['中等职业教育专任教师数_市辖区'].tolist())
        .add_yaxis('中等职业教育在校生数_市辖区',data['中等职业教育在校生数_市辖区'].tolist(),is_smooth=True)
        .set_global_opts(title_opts=opts.TitleOpts(title="教育细分指标"))
    )
    print(data.columns)
    return c
line_base(edudata).render_notebook()
```
