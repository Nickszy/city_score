```python
import pandas as pd 
# 数据库
from sqlalchemy import create_engine
engine = create_engine('mysql://root:root@localhost:3306/citylist?charset=utf8')
# stock_basic.to_sql('stock_basic',engine,index=False,if_exists='append')
```

```python
import os
f = [c for c in os.listdir('../长三角城市年度数据')]
```

```python
df =pd.DataFrame()
for file in f:
    data = pd.read_excel('../长三角城市年度数据/'+file)
    localname = data.iloc[0,1]
    data['name']= localname
    data.drop(index=[0,1,2],inplace=True)
    col_name = data.columns.values
    col_name[0] = 'year'
    data.columns = col_name
    print(data.shape,localname)
    data.to_sql('city_data2',engine,index=False,if_exists='append')
```
